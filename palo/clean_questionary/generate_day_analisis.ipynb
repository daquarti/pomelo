```python
def generate_lista_days (data):
    days_questionary = pd.to_datetime(data.index)
    days_before = []
    for x in days_questionary:
        before = x - pd.Timedelta(days=2)
        days_before.append(before)
    lista_days = list (zip(days_before, days_questionary))
    return lista_days
    
```

```python
def generate_dataframe (lista_days, path):

    lista_to_concat = []

    for x in lista_days:

        df = pd.read_csv(path)


        df['dates'] = pd.to_datetime(df.Date)

        df = df[(df.dates > x[0]) & (df.dates < x[1])]

        dia = x[1]

        df = df[['kcals', 'Average Hourly kcals', 'METs', 'Number of Freedson (1998) Bouts occurring in this day', 'Total time of Freedson (1998) Bouts occurring in this day', 'Total activity counts of Freedson (1998) Bouts occurring in this day', 'Number of Sedentary Bouts occurring in this day', 'Total time of Sedentary Bouts occurring in this day', 'Number of Sedentary Breaks occurring in this day', 'Total time of Sedentary Breaks occurring in this day', '% in Sedentary', '% in Light', '% in Moderate', '% in Vigorous',
               '% in Very Vigorous', 'Total MVPA', '% in MVPA', 'Average MVPA Per Hour', 'Axis 1 Counts', 'Axis 2 Counts', 'Axis 3 Counts', 'Axis 1 Average Counts', 'Axis 2 Average Counts', 'Axis 3 Average Counts', 'Axis 1 Max Counts', 'Axis 2 Max Counts', 'Axis 3 Max Counts', 'Axis 1 CPM', 'Axis 2 CPM', 'Axis 3 CPM', 'Vector Magnitude Counts', 'Vector Magnitude Average Counts', 'Vector Magnitude Max Counts', 'Vector Magnitude CPM', 'Steps Counts', 'Steps Average Counts', 'Steps Max Counts', 'Steps Per Minute', 'Lux Average Counts', 'Lux Max Counts', 'Number of Epochs']]

        df = df.reset_index()

        v = df.unstack().to_frame().sort_index(level=1).T

        index_0 = [str (x) for x in v.columns.droplevel(0)]

        index_1 = v.columns.droplevel(1).to_list()

        lista_zip = []
        for x in list(zip(index_0, index_1)):
            name  = x[1] + "_previus_day_" + x[0]
            lista_zip.append (name)
        v.columns = lista_zip
        v['date'] = dia
        lista_to_concat.append(v)
    final = pd.concat(lista_to_concat)
    final = final.set_index('date')
    final.index = final.index.astype(str)
    return final
    
```
