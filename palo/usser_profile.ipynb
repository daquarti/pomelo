```python
import pandas as pd
```

```python
df = pd.read_csv ('questionnaire data/user_profiles_lupa.csv')
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df.columns
```

```python
df["datetime"] = pd.to_datetime(df["created_at"])
```

```python
df["datetime_up"] = pd.to_datetime(df["updated_at"])
```

```python
y = df["datetime_up"] - df["datetime"]
```

```python
df['segundos_activo'] = y.dt.seconds
```

```python
segundos_activos = df[['user_id', 'segundos_activo']]
```

```python
segundos_activos.to_csv('segundos_activos.csv')
```
